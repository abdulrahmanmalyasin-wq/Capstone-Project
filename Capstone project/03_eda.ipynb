```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Data
df= pd.read_csv("C:\\Users\\MOJ\\Desktop\\semester 2\\Capstone project\\data\\clean_data.csv")
```

```python
#copying the data 
df_copy = df.copy()
```

```python
# Histograms / KDE for 3 numerical features

# 1. SalePrice
plt.figure()
sns.histplot(df_copy['SalePrice'], kde=True)
plt.title("SalePrice Distribution")
plt.show()
# right skewed distribution
 
# 2. Lot Area
plt.figure()
sns.histplot(df_copy['Lot Area'], kde=True)
plt.title("Lot Area Distribution")
plt.show()
# right skewed distribution

# 3. Overall Qual
plt.figure()
sns.histplot(df_copy['Overall Qual'], kde=True)
plt.title("Overall Qual Distribution")
plt.show()
# symmetric distribution
```

```python
# compare SalePrice across 2 categories

plt.figure()
sns.boxplot(x=df_copy['Overall Qual'], y=df_copy['SalePrice'])
plt.title("SalePrice vs Overall Qual")
plt.show()

plt.figure()
sns.boxplot(x=df_copy['Neighborhood'], y=df_copy['SalePrice'])
plt.title("SalePrice vs Neighborhood")
plt.xticks(rotation=45)
plt.show()
```

```python
# top 10 features most correlated with SalePrice

corr = df_copy.corr(numeric_only=True)['SalePrice'].abs().sort_values(ascending=False)

top10 = corr.head(10).index

plt.figure()
sns.heatmap(df_copy[top10].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Top 10 Features Correlated with SalePrice")
plt.show()
```

```python
# key relationship with a third variable (color)

plt.figure()
sns.scatterplot(
    x=df_copy['Gr Liv Area'],
    y=df_copy['SalePrice'],
    hue=df_copy['Overall Qual']
)
plt.title("Gr Liv Area vs SalePrice (colored by Overall Qual)")
plt.show()
```

```python
# Groupby summary: mean SalePrice by category

group_mean = df_copy.groupby('Neighborhood')['SalePrice'].mean().sort_values()

print(group_mean)
```
