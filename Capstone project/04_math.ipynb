```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import  StandardScaler,OrdinalEncoder
```

```python
# Load the Data
df= pd.read_csv("C:\\Users\\MOJ\\Desktop\\semester 2\\Capstone project\\data\\clean_data.csv")
```

```python
#copying the data 
df_copy = df.copy()
```

```python
# Compute mean and standard deviation of SalePrice manually using NumPy

values = df_copy['SalePrice'].values
mean_saleprice = np.sum(values) / len(values) # Calculate mean
std_saleprice = np.sqrt(np.sum((values - mean_saleprice) ** 2) / len(values)) # Calculate standard deviation

print("Mean SalePrice:", mean_saleprice)
print("Std SalePrice:", std_saleprice)
```

```python
# Standardise one column manually and compare it with StandardScaler

# choose column
values = df_copy['Gr Liv Area'].values

# manual mean and std
mean = np.mean(values)
std = np.std(values)

# manual z-score (broadcasting)
df_copy['Gr_Liv_Area_manual'] = (values - mean) / std

# StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_copy['Gr_Liv_Area_scaled'] = scaler.fit_transform(df_copy[['Gr Liv Area']])

# compare
print(df_copy[['Gr_Liv_Area_manual', 'Gr_Liv_Area_scaled']].head())
```

```python
# Compute cosine similarity using the given function

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# get indices of highest and lowest SalePrice
idx_max = df_copy['SalePrice'].idxmax()
idx_min = df_copy['SalePrice'].idxmin()

# select numeric features only (exclude target) + handle missing values
features = df_copy.select_dtypes(include=['number']).drop(columns=['SalePrice']).fillna(0)

# extract feature vectors
vec_max = features.loc[idx_max].values
vec_min = features.loc[idx_min].values

# compute cosine similarity
cos_sim = cosine_similarity(vec_max, vec_min)

print("Cosine Similarity:", cos_sim)
```

```python
# Estimate probability using conditional probability P(A|B)

threshold = np.percentile(df_copy['SalePrice'], 75)

high_price = df_copy['SalePrice'] > threshold
high_quality = df_copy['Overall Qual'] >= 8

both_conditions = high_price & high_quality

probability = both_conditions.sum() / high_quality.sum()

print("Probability:", probability)
```
