```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import  StandardScaler,OrdinalEncoder
```

```python
# Load the Data
df= pd.read_csv("C:\\Users\\MOJ\\Desktop\\semester 2\\Capstone project\\data\\clean_data.csv")
```

```python
#copying the data 
df_copy = df.copy()
```

```python
# Compute mean and standard deviation of SalePrice manually using NumPy

values = df_copy['SalePrice'].values
mean_saleprice = np.sum(values) / len(values) # Calculate mean
std_saleprice = np.sqrt(np.sum((values - mean_saleprice) ** 2) / len(values)) # Calculate standard deviation

print("Mean SalePrice:", mean_saleprice)
print("Std SalePrice:", std_saleprice)
```

```python

# Compute mean and standard deviation manually + standardize one column and compare

# manual mean and std
values = df_copy['SalePrice'].values
mean_saleprice = np.sum(values) / len(values)
std_saleprice = np.sqrt(np.sum((values - mean_saleprice) ** 2) / len(values))

print("Mean SalePrice:", mean_saleprice)
print("Std SalePrice:", std_saleprice)

df_copy['Gr_Liv_Area_manual'] = (values - mean_saleprice) / std_saleprice

# StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_copy['Gr_Liv_Area_scaled'] = scaler.fit_transform(df_copy[['Gr Liv Area']])

# compare
print("\nManual Standardization:")
print(df_copy[['Gr_Liv_Area_manual', 'Gr_Liv_Area_scaled']].head())
```

```python
# Compute cosine similarity between highest and lowest SalePrice feature vectors

# get indices of highest and lowest SalePrice
idx_max = df_copy['SalePrice'].idxmax()
idx_min = df_copy['SalePrice'].idxmin()

# select numeric features only (exclude target)
features = df_copy.select_dtypes(include=['number']).drop(columns=['SalePrice'])

# extract feature vectors
vec_max = features.loc[idx_max].values
vec_min = features.loc[idx_min].values

# compute cosine similarity manually
dot_product = np.dot(vec_max, vec_min)
norm_max = np.linalg.norm(vec_max)
norm_min = np.linalg.norm(vec_min)

cosine_similarity = dot_product / (norm_max * norm_min)

print("Cosine Similarity:", cosine_similarity)
```

```python
# Estimate probability: fraction of high-quality houses with SalePrice above a threshold

threshold = df_copy['SalePrice'].quantile(0.75) # Set threshold at 75%

high_quality = df_copy[df_copy['Overall Qual'] >= 8] # Filter high quality houses

prob = (high_quality['SalePrice'] > threshold).mean() #calculate probability

print("Probability:", prob)
```
